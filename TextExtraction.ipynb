```python
import spacy
import csv
import re

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    # Process the text using spaCy
    doc = nlp(text)

    # Extract RFQ Number, RFQ Date, GST Number, PAN, Order Currency, Delivery Place
    rfq_number = re.search(r'RFQ Number\s*:\s*([\w\d]+)', text).group(1)
    rfq_date = re.search(r'RFQ Date\s*:\s*([\d.]+)', text).group(1)
    gst_number = re.search(r'GST Number\s*:\s*([\w\d]+)', text).group(1)
    pan = re.search(r'PAN\s*:\s*([\w\d]+)', text).group(1)
    order_currency = re.search(r'Order Currency\s*:\s*([\w]+)', text).group(1)
    delivery_place = re.search(r'Delivery Place\s*:\s*([\w\s]+)', text).group(1)

    # Extract Vendor Details
    vendor_details_match = re.search(r'Vender Address:(.+?)Kind Attention', text, re.DOTALL)
    vendor_details = vendor_details_match.group(1).strip() if vendor_details_match else ''

    # Extract Material/Asset/Service Details
    material_details_match = re.search(r'SI\.No\..+?(\d+\s+.+?)Terms & Conditions:', text, re.DOTALL)
    material_details = material_details_match.group(1).strip() if material_details_match else ''

    return rfq_number, rfq_date, gst_number, pan, order_currency, delivery_place, vendor_details, material_details

# Example text
quotation_text = """
REQUEST FOR QUOTATION

Vendor Details RFQ Number     : 2000014380

RFQ Date           : 14.06.2023

GST Number      : 2QAAATRO7S8A1ZP

PAN                    : AAATRO758A

Order Currency   : INR

Delivery Place     : R.V. College of Engineering

Details

Vender Address:

Starcom Information

Technology Ltd.

NO.88 Times Square

M.G. Road

Bengaluru-560001

Karnataka

Vendor Code :10000226

GST Number

29AAACJI380RIZI1

PAN: AAACJ1380R

Kind Attention

Dear Sir / Madam,



You are requested to quote for the supply of the below mentioned items on or before 22.06.2023.

SI.No. Material / Asset / Service

Description / Specification

HSN Code /

SAC

Quantity UOM

1 60004342

Raspberry Pi4

 10 No

2 60000067

Arduino UNO

 10 EA

Terms & Conditions:



You are requested to quote for the same furnishing complete details, along with brochures, warranty etc...

The sealed quotation should be addressed to "The Principal, RV College of Engineering. Mysore Road, Bengaluru-

560059" and quotation should be sent through post or courier only.

Kindly specify on the Envelope and Quotation Kind Attn: HoD, Electronics and Telecommunication Engineering.



Specifications:



1.Raspberry P14, 8GB RAM, Desktop Computer Integrated with 64Bit Quad Core Cortex-A72ARM V8, Board

Com BCM2711. Bluetooth, Ethernet, Wireless at 2.4Ghz & 5Ghz, USB 3.0 Ports, HDMI Port, Original PI Adaptor

& Cables/Connectors, 64GB Class SD Card to load OS & Data storage



2.Arduino UNO R3, ATmega328P, USB Cable, Power adaptor.

 For: RV College of Engineer







           Authorized Signatory
"""

# Extract entities from the text
rfq_number, rfq_date, gst_number, pan, order_currency, delivery_place, vendor_details, material_details = extract_entities(quotation_text)

# Store entities in a CSV file
csv_data = [
    ('RFQ Number', 'RFQ Date', 'GST Number', 'PAN', 'Order Currency', 'Delivery Place', 'Vendor Details', 'Material Details'),
    (rfq_number, rfq_date, gst_number, pan, order_currency, delivery_place, vendor_details, material_details)
]

csv_file_path = 'extracted_entities.csv'
with open(csv_file_path, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerows(csv_data)

# Print extracted entities
print("RFQ Number:", rfq_number)
print("RFQ Date:", rfq_date)
print("GST Number:", gst_number)
print("PAN:", pan)
print("Order Currency:", order_currency)
print("Delivery Place:", delivery_place)
print("Vendor Details:", vendor_details)
print("Material Details:", material_details)
print("Entities stored in CSV:", csv_file_path)
```

```python
import re
import csv

def extract_features(text):
    # Extract features using regular expressions
    vendor_name_match = re.search(r'Vender Name\s*:\s*([\w\s]+)', text)
    vendor_name = vendor_name_match.group(1).strip() if vendor_name_match else ''

    vendor_address_match = re.search(r'Vender Address\s*:\s*([\w\s\.,-]+)', text)
    vendor_address = vendor_address_match.group(1).strip() if vendor_address_match else ''

    vendor_gstin_match = re.search(r'Vender GSTIN/UIN\s*:\s*([\w\d]+)', text)
    vendor_gstin = vendor_gstin_match.group(1).strip() if vendor_gstin_match else ''

    vendor_pan_match = re.search(r'Vender PAN\s*:\s*([\w\d]+)', text)
    vendor_pan = vendor_pan_match.group(1).strip() if vendor_pan_match else ''

    buyer_name_match = re.search(r'Buyer Name\s*:\s*([\w\s]+)', text)
    buyer_name = buyer_name_match.group(1).strip() if buyer_name_match else ''

    buyer_address_match = re.search(r'Buyer Address\s*:\s*([\w\s\.,-]+)', text)
    buyer_address = buyer_address_match.group(1).strip() if buyer_address_match else ''

    invoice_number_match = re.search(r'Invoice Number\s*:\s*([\w\d]+)', text)
    invoice_number = invoice_number_match.group(1).strip() if invoice_number_match else ''

    date_match = re.search(r'Date\s*:\s*([\d.]+)', text)
    date = date_match.group(1).strip() if date_match else ''

    description_match = re.search(r'Description of Goods/Items\s*:\s*([\w\s\.,-]+)', text)
    description = description_match.group(1).strip() if description_match else ''

    hsn_sac_match = re.search(r'HSN/SAC codes\s*:\s*([\w\d]+)', text)
    hsn_sac = hsn_sac_match.group(1).strip() if hsn_sac_match else ''

    quantity_match = re.search(r'Quantity\s*:\s*([\w\d]+)', text)
    quantity = quantity_match.group(1).strip() if quantity_match else ''

    rate_match = re.search(r'Rate per unit\s*:\s*([\w\d.]+)', text)
    rate = rate_match.group(1).strip() if rate_match else ''

    total_amount_match = re.search(r'Total amount\s*:\s*([\w\d.]+)', text)
    total_amount = total_amount_match.group(1).strip() if total_amount_match else ''

    taxable_value_match = re.search(r'Taxable value\s*:\s*([\w\d.]+)', text)
    taxable_value = taxable_value_match.group(1).strip() if taxable_value_match else ''

    cgst_match = re.search(r'CGST rate and amount\s*:\s*([\w\d.%]+)', text)
    cgst = cgst_match.group(1).strip() if cgst_match else ''

    sgst_utgst_match = re.search(r'SGST/UTGST rate and amount\s*:\s*([\w\d.%]+)', text)
    sgst_utgst = sgst_utgst_match.group(1).strip() if sgst_utgst_match else ''

    total_tax_amount_match = re.search(r'Total tax amount\s*:\s*([\w\d.]+)', text)
    total_tax_amount = total_tax_amount_match.group(1).strip() if total_tax_amount_match else ''

    total_payable_amount_match = re.search(r'Total payable amount\s*:\s*([\w\d.]+)', text)
    total_payable_amount = total_payable_amount_match.group(1).strip() if total_payable_amount_match else ''

    amount_in_words_match = re.search(r'Amount in words\s*:\s*([\w\s.,-]+)', text)
    amount_in_words = amount_in_words_match.group(1).strip() if amount_in_words_match else ''

    terms_conditions_match = re.search(r'Terms & Conditions\s*:\s*([\w\s.,-]+)', text)
    terms_conditions = terms_conditions_match.group(1).strip() if terms_conditions_match else ''

    authorized_signatory_match = re.search(r'Authorized Signatory\s*:\s*([\w\s.,-]+)', text)
    authorized_signatory = authorized_signatory_match.group(1).strip() if authorized_signatory_match else ''

    return {
        "Vendor Name": vendor_name,
        "Vendor Address": vendor_address,
        "Vendor GSTIN/UIN": vendor_gstin,
        "Vendor PAN": vendor_pan,
        "Buyer Name": buyer_name,
        "Buyer Address": buyer_address,
        "Invoice Number": invoice_number,
        "Date": date,
        "Description of Goods/Items": description,
        "HSN/SAC codes": hsn_sac,
        "Quantity": quantity,
        "Rate per unit": rate,
        "Total amount": total_amount,
        "Taxable value": taxable_value,
        "CGST rate and amount": cgst,
        "SGST/UTGST rate and amount": sgst_utgst,
        "Total tax amount": total_tax_amount,
        "Total payable amount": total_payable_amount,
        "Amount in words": amount_in_words,
        "Terms & Conditions": terms_conditions,
        "Authorized Signatory": authorized_signatory
    }

# Example text
invoice_text = """
REQUEST FOR QUOTATION

Vendor Details RFQ Number     : 2000014380

RFQ Date           : 14.06.2023

GST Number      : 2QAAATRO7S8A1ZP

PAN                    : AAATRO758A

Order Currency   : INR

Delivery Place     : R.V. College of Engineering

Details

Vender Address:

Starcom Information

Technology Ltd.

NO.88 Times Square

M.G. Road

Bengaluru-560001

Karnataka

Vendor Code :10000226

GST Number

29AAACJI380RIZI1

PAN: AAACJ1380R

Kind Attention

Dear Sir / Madam,



You are requested to quote for the supply of the below mentioned items on or before 22.06.2023.

SI.No. Material / Asset / Service

Description / Specification

HSN Code /

SAC

Quantity UOM

1 60004342

Raspberry Pi4

 10 No

2 60000067

Arduino UNO

 10 EA

Terms & Conditions:



You are requested to quote for the same furnishing complete details, along with brochures, warranty etc...

The sealed quotation should be addressed to "The Principal, RV College of Engineering. Mysore Road, Bengaluru-

560059" and quotation should be sent through post or courier only.

Kindly specify on the Envelope and Quotation Kind Attn: HoD, Electronics and Telecommunication Engineering.



Specifications:



1.Raspberry P14, 8GB RAM, Desktop Computer Integrated with 64Bit Quad Core Cortex-A72ARM V8, Board

Com BCM2711. Bluetooth, Ethernet, Wireless at 2.4Ghz & 5Ghz, USB 3.0 Ports, HDMI Port, Original PI Adaptor

& Cables/Connectors, 64GB Class SD Card to load OS & Data storage



2.Arduino UNO R3, ATmega328P, USB Cable, Power adaptor.

 For: RV College of Engineer







           Authorized Signatory"""

# Extract features
features = extract_features(invoice_text)

# Store features in a CSV file
csv_file_path = 'extracted_features1.csv'
with open(csv_file_path, 'w', newline='') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=features.keys())
    csv_writer.writeheader()
    csv_writer.writerow(features)

print("Features stored in CSV file:", csv_file_path)
```
